```python
import pandas as pd
import numpy as np


df = pd.read_csv('../model/roberta_wwm_large_512_1_last3embedding_cls/roberta_wwm_large_512_1_last3embedding_cls_0/sub.csv')
df = df[['id']]
df['0'] = 0
df['1'] = 0
df['2'] = 0

k = 5
for i in [0,1,2,3,4]:
    temp=pd.read_csv('../model/roberta_wwm_large_512_1_last3embedding_cls/roberta_wwm_large_512_1_last3embedding_cls_{}/sub.csv'.format(i))
    df['0']+=temp['label_0']/k
    df['1']+=temp['label_1']/k
    df['2']+=temp['label_2']/k
print(df['0'].mean())

df['label']=np.argmax(df[['0','1','2']].values,-1)
df[['id','label']].to_csv('../model/roberta_wwm_large_512_1_last3embedding_cls/roberta_wwm_large_512_1_last3embedding_cls_sub.csv',index=False)
```
